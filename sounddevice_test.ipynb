Choose sounddevice library to record sound<br>
Out to numpy array + understandable usage

```python
import sounddevice as sd
import os

import scipy.io.wavfile
import matplotlib.pyplot as plt
```

Play:

```python
file_name = "why_we_get_mad_1.wav"
file_path = os.path.join("data", file_name)
```

```python
samplerate, waveform = scipy.io.wavfile.read(file_path)
```

```python
sd.play(waveform)
```

```python
sd.stop()
```

```python
plt.plot(waveform)
plt.show()
```

Record:

```python
duration = 5 #seconds
fs = 44100
sd.default.samplerate = fs
sd.default.channels = 1
```

```python
myrecording = sd.rec(int(duration * fs))
```

```python
sd.wait()
print(myrecording)
```

```python
sd.play(myrecording)
```

```python
sd.wait()
```

```python
plt.plot(myrecording)
plt.show()
```
